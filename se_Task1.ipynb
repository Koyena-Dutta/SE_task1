```python
#multiple sets using reading a file
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temp_model(time, a, b, c):
    # Quadratic equation: t = at^2 + bt + c
    temperature = a * (time**2) + b * time + c
    return temperature

def read_coefficients_from_file(file_path):
    coefficient_sets = []
    with open(file_path, 'r') as file:
        for line in file:
            coefficients = list(map(float, line.split()))
            coefficient_sets.append(coefficients)
    return coefficient_sets

def main():
    file_path = 'coefficients.txt'
    coefficient_sets = read_coefficients_from_file(file_path)
    time_values = np.linspace(0, 10, 51)

    # Plotting each set of coefficients
    for i, coefficients in enumerate(coefficient_sets):
        a, b, c = coefficients
        temperature_values = quadratic_temp_model(time_values, a, b, c)
        label = f"Set {i+1} - {a , b , c}"
        plt.plot(time_values, temperature_values, label=label)

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title("Multiple Sets Weather Modeling")
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#reading a file
import matplotlib.pyplot as plt
import numpy as np

def quadratic_temp_model(time, a, b, c):
    # Quadratic equation: t = at^2 + bt + c
    temperature = a * (time**2) + b * time + c
    return temperature

def main():
    # Read coefficients from the file
    try:
        with open("weather.txt", "r") as file_object:
            lines = file_object.readlines()
            a = float(lines[0].strip())
            b = float(lines[1].strip())
            c = float(lines[2].strip())
    except FileNotFoundError:
        print("Error: File 'weather.txt' not found.")
        return
    except ValueError:
        print("Error: Invalid data in 'weather.txt'. Expected three lines with numeric coefficients.")
        return

    # Time values from 0 to 10 with step 0.2
    time_values = np.linspace(0, 10, 51)

    # Calculate temperatures using coefficients
    temperature_hardcoded = quadratic_temp_model(time_values, a, b, c)

    # Plotting
    plt.plot(time_values, temperature_hardcoded, label="user input coeff")
    plt.xlabel('time')
    plt.ylabel('temperature')
    plt.legend()
    plt.title(" Weather Modeling using file")
    plt.show()

if __name__ == "__main__":
    main()
```
